```python
#variable length arguments
#positional argument
#keywords arguments

#*args  positional arguments and variable

def print_numbers(*args):
    for numbers in args:
        print(numbers)


print_numbers(1,2,3,4,5,6,"Nikhil")
# these are all the positional arguments
```

```python
#keywords argument
# it is denoted by **
# in keywords argument all the paramenters are in key value pair
def print_detail(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")


print_detail(name="Nikhil",age="32")
```

```python
# argument does not have key value it is a positional argument 
# argument does have key value it is a keyword argument 
```

```python
def print_both(*args,**kwargs):
    for val in args:
        print(f"Positional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

# positional argument cannot comes after keywods argument
#print_both(1,2,3,4,name="Nikhil",age="24","Rathod") # it is wrong
print_both(1,2,3,4,"Rathod", name="Nikhil", age="24") # it is right
```

```python
#example 1 temp conversion
def temp_convert(temp,unit):
    if unit=="C":
        return temp * (9/5) +32 # cel to fahrenheit

    elif unit == "F":
        return (temp-32)*5/9
    
    else:
        return  None

print(temp_convert(25,"C"))
print(temp_convert(25,"F"))

```

```python
#Example 3 - Calculate the total cost of item in shopping cart

def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']*item['quantity']
    return total_cost
        
        


cart=[
    {'name':'Apple','price':0.5,'quantity':4},
    {'name':'Banana','price':0.3,'quantity':6},
    {'name':'orange','price':0.7,'quantity':3}
]

total_cost=calculate_total_cost(cart)
print(total_cost)
```
