# Sets

####  sets are built-in data type python used for store collection of unique items they are unordered meaning that the element do not follow a specific order and they do not allow duplicate element 

```python
# creating a set

my_set={1,2,3,4}
print(type(my_set))
print(my_set)
```

```python
my_empty_set={}
my_1_empty_set=set()
print(my_empty_set)
print(my_1_empty_set)
```

```python
# inside set we can give list tuple 

set_1=set([1,2,3,4])
print(set_1)
print(type(set_1))
```

```python
my_set=set({1,2,3,4,5})
print(my_set)
print(type(my_set))
```

```python
my1_set=set([1,5,9,9,3,4,4,4,5,6,7,7])
print(my1_set)
print(type(my1_set))
```

```python
#Basic set operation 

my1_set.add(10)
print(my1_set)
```

```python
my1_set.remove(9)
print(my1_set)
```

```python
my1_set.remove(11)  # Remove an element from a set; it must be a member.
print(my1_set)
```

```python
my1_set.discard(12)  
#Remove an element from a set if it is a member.
#Unlike set.remove(), the discard() method does not raise
#an exception when an element is missing from the set.
print(my1_set)
```

```python
#pop method 
#it works on first in first out 
# print(my1_set)
removed_element=my1_set.pop()
print(removed_element)
print(my1_set)
```

```python
my1_set.clear()
print(my1_set)
```

```python
#set membership test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
# mathematical operation
set_1={1,2,3,4,5,6}
set_2={4,5,6,7,8,9}
# union 
union_set=set_1.union(set_2)
print(union_set)

# Intersection

intersection_set=set_1.intersection(set_2)
print(intersection_set)

# here it will find out common element and update the set_1  perform that operation and update
set_1.intersection_update(set_2)
print(set_1)
```

```python

#difference operation
#common element is removed
set_3={1,2,3,4,5,6}
set_4={4,5,6,7,8,9}
print(set_3.difference(set_4))  # it only look for element in set_3 not in set_4

print(set_4.difference(set_3))  # sets are unordered
```

```python
# symmetric difference
# it will remove the common element and print the uncommon element from both the sets

print(set_3.symmetric_difference(set_4)) 
print(set_4.symmetric_difference(set_3)) 
```

```python
# sets methods
#  a.subset(b)  "a in inside b"
#  a.issuperset(b) "a contains b"
set_5={1,2,3,4,5}
set_6={3,4,5}
print(set_5.issubset(set_6)) # it checks that set_5 element present in set_6 or not it will give bollean 
print(set_5.issuperset(set_6)) # it checks set_5 contain all the element of set_6 or not it will give boolean
```

```python
lst=[1,2,2,3,4,4,5]
set(lst)
```

```python
# counting uniaue elements in text

text="In this tutorial we  are discussing about sets"
words=text.split()

# convert list of sets to get unique words
unique_words=set(words)

print(unique_words)
print(len(unique_words))


```
