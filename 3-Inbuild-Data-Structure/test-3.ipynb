### Slicing List

```python
numbers=[1,2,3,4,5,6,7,8]
print(numbers[2:6])
print(numbers[:6])
print(numbers[5:])
print(numbers[1:2:7]) #start:stop:step
print(numbers[::2])
print(numbers[::-1]) # it will return reverse list
```

```python
# Iterating over list

for i in numbers:
    print(i)
```

```python
# iterating with index
for index,number in enumerate(numbers):
    print(index,number)
```

```python
# list Comprehension
lst=[]
for x in range(10):
    lst.append(x**2)
print(lst)
```

```python
#No, list comprehension does NOT use lambda functions internally.
#List Comprehension = Built-in Python syntax (like for loops)
new_lst=[x**2 for x in range(10)]
print(new_lst)
```

```python
#basic syntax [expression for item in iterable]
# with conditional logic [expression for item in iterable if condition]

square=[num**2 for num in range(10)]
print(square)
```

```python
# list comprehenssion with condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)
    print(lst)

```
