```python
# Nested Dictionaries
students={
    "student1":{"name":"Nikhil","age":"26"},
    "student2":{"name":"Lavish","age":"23"},
    "student3":{"name":"Aditya","age":"31"}
}
print(students)
```

```python
for key ,value in students.items():
    print(key,":",value)
```

```python
# Access the nexted dictionaries elements
print(students["student1"]["name"])
print(students["student2"]["name"])
print(students["student3"]["name"])
print(students["student1"]["age"])
```

```python
# iterating over dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")

```

```python
# Dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
#conditional dictionary comprehension
even_square={x:x**2 for x in range(10) if x%2==0}
print(even_square)
```

```python
#practical example
# use a dictionary to count the frequency of element in list
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
    
```

```python
# merge to dictionary in one 
#**dict unpacks all key-value pairs
# Right side wins in conflicts (dict2 overrides dict1)
dict1={"a":"1","b":"2"}
dict2={"b":"3","c":"4"}
merge_dict={**dict1,**dict2}
print(merge_dict)
```
