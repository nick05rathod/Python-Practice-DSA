```python
#if statement
#if statement check the condition and execute the code if the condition is true
#the else statement will execute block of code if the condition is false
age=20
if age>=18:
    print("You can drive")
else:
    print("you cannot drive")

```

```python
#elif statement is used to check multiple conditions
age=20
if age<13:
    print("you are child")
elif age<18:
    print("Your are teeneager")
else:
    print("your are an adult")
```

```python
# nested if statement is used to check multiple conditions within another condition

num=int(input("Enter your number: "))
if num>0:
    if num%2==0:
        print("The number is positive and even")
    else:
        print("The number is positive and odd")
else:
    print("The number is not positive")
```

```python
# practical example of if statement
year=int(input("Enter the year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print(f"The {year} is leap year")
else:
    print(f"The {year} is not leap year")   
```
