```python
#if statement
#if statement check the condition and execute the code if the condition is true
#the else statement will execute block of code if the condition is false
age=20
if age>=18:
    print("You can drive")
else:
    print("you cannot drive")

```

```python
#elif statement is used to check multiple conditions
age=20
if age<13:
    print("you are child")
elif age<18:
    print("Your are teeneager")
else:
    print("your are an adult")
```

```python
# nested if statement is used to check multiple conditions within another condition

num=int(input("Enter your number: "))
if num>0:
    if num%2==0:
        print("The number is positive and even")
    else:
        print("The number is positive and odd")
else:
    print("The number is not positive")
```

```python
# practical example of if statement
year=int(input("Enter the year: "))
if (year%4==0 and year%100!=0) or (year%400==0):
    print(f"The {year} is leap year")
else:
    print(f"The {year} is not leap year")   
```

```python
leap_year=int(input("Enter the year: "))
if leap_year%4==0:
    if leap_year%100==0:
        if leap_year%400==0:
            print(f"The {leap_year} is Leap year")
        else:
            print(f"The {leap_year} is not leap year")
    else:
        print(f"The {leap_year} is leap year")
else:
    print(f"The {leap_year} is not leap year")
```

## While loop

#### while loop is continuous to execute as loong as condition is ture

```python
count=0
while count<5:
    print(count)
    count+=1
```

```python
count=0
while count%2==0:
    print(count)
    count+=1
```

```python
# loop control statement
# 1. Break - The statement exits the loop prematurly

for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
 # 2. Continue
# The statement skips the current iteration and continue with the next 

for i in range(0,10):
    if i==5:
        continue
    print(i)
```

```python
# it will print odd numbers
for i in range(0,10):
    if i%2==0:
        continue
    print(i)
```

```python
# 3. Pass
# the pass statement is a null operator, it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
```

## Nested Loops
#### Loop insode a loop

```python
for i in range(3):
    for j in range(2):
        print(f"i={i} and j={j}")

# total 3*2= 6 iterations
```

```python
 # calculate the sum of first n natural number using while and for loop


num=10
sum=0
count=1
while  count<=num:
    sum=sum+count
    count+=1

print("Sum of first of n natural number: ",sum)
```
