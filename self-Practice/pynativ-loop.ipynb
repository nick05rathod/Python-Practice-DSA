```python
# count voels and consonant in a sentence
# isalpha() checks if a character/string contains only alphabetic(a-z, A-Z)
# it returns ture i all  character are letter , Flase otherwise 
my_input="Loops are fun"
vowel="aeiou"
vowel_count=0
cons_count=0
for char in my_input.lower():
    if char.isalpha():
        if char in vowel:
            vowel_count+=1
        else:
            cons_count+=1
print("vowel: ", vowel_count)
print("consonant: ", cons_count)
```
